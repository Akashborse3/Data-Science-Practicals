```python
import random
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from warnings import filterwarnings
filterwarnings("ignore")
```

```python
# It takes the random values for column
Math_Score = [random.randint(65, 85) for i in range(250)]


# For importing the missing values
missing_pct = 0.2  # 20% missing values

# Calculate the number of missing values
n_missing = int(len(Math_Score) * missing_pct)

# Generate a list of indices to replace with missing values
missing_indices = random.sample(range(len(Math_Score)), n_missing)

# Replace the values at the missing indices with None
for i in missing_indices:
    Math_Score[i] = None
```

## Finding the IQR(Inter Quartile Range)

```python
percentile25 = DF['Math_Score'].quantile(0.25)
percentile75 = DF['Math_Score'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
IQR = percentile75 - percentile25
IQR
```

```python
Upper_Limit = percentile75 + 1.5 * IQR
Lower_Limit = percentile25 - 1.5 * IQR

print("Upper limit",Upper_Limit)
print("Lower limit",Lower_Limit)
```

## Finding Outliers

```python
DF[DF['Math_Score'] > Upper_Limit]
```

```python
DF[DF['Math_Score'] < Lower_Limit]
```

# Trimming

* Trimming is a method of outlier detection that involves removing a certain percentage of the highest and lowest values in a dataset before calculating the mean or median. The percentage of values to be trimmed is usually chosen based on the characteristics of the data and the desired level of outlier robustness.

```python
New_DF = DF[DF['Math_Score'] > Lower_Limit]
New_DF.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(DF['Math_Score'])

plt.show()
```

## Capping

* Capping is a method of outlier detection that involves replacing extreme values in a dataset with a less extreme value, typically the next highest or lowest value within a certain range. 

```python
New_DF_Cap = DF.copy()

New_DF_Cap['Math_Score'] = np.where(
    New_DF_Cap['Math_Score'] > Upper_Limit,
    Upper_Limit,
    np.where(
        New_DF_Cap['Math_Score'] < Lower_Limit,
        Lower_Limit,
        New_DF_Cap['Math_Score']
    )
)
```
